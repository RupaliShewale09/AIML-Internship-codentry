```python
# 2. Download dataset (cached)
# ===============================
import kagglehub

cache_path = kagglehub.dataset_download(
    "muki2003/yolo-drone-detection-dataset"
)

print("Cached dataset path:", cache_path)
```

```python
# 3. Copy dataset to required folder
# ===============================
import os
import shutil

TARGET_DIR = "/content/datasets/yolo_drone"

# Create base directory
os.makedirs("/content/datasets", exist_ok=True)

# Copy only once
if not os.path.exists(TARGET_DIR):
    shutil.copytree(cache_path, TARGET_DIR)
    print("Dataset copied successfully!")
else:
    print("Dataset already exists!")
```

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/datasets/yolo_drone/drone_dataset/data.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    device=0,        # GPU
    workers=2,
    name="drone_yolov8"
)
```
