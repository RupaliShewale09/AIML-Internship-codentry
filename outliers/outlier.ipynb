```python
# MEAN = average
age_mean = data["age"].mean().round()
print(age_mean)
```

```python
# MODE = most repeated
age_mode = data["age"].mode()[0]
print(age_mode)
```

```python
# MEDIAN = middle value when data is arranged in order
age_median = data["age"].median()
print(age_median)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data["age"])

plt.title("Box Plot for Outliers Detection")
plt.xticks(rotation=30)
# plt.yticks(np.arange(0, max(data["attendance"])+5, 5))
plt.ylim(10, 65) 
plt.show()
```
